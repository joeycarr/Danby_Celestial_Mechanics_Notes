# 1.2 Some Definitions

 - revolve: motion around a point
 - rotate: motion about an axis
 - orbit: the course of a planet, comet, satellite, etc.
two-body-problem: the motion of two point masses which move subject only to their mutual gravitational attraction
perturbation: deviations from a two-body description of an orbit caused by the added attraction of other bodies

# 1.3 Orbital Definitions

 - ap-: point in orbit of greatest separation, e.g. "apogee", "apastron", or "aphelion"
 - peri-: point in orbit of least separation, e.g. "perigee", "periastron", or "perihelion"
 - sidereal period: the time taken for a complete revolution in an orbit, with respect to the stars
 - synodic period: the mean synodic period is the average time taken for a planet to return to the same position relative to the Earth, e.g. the time between one opposition and the next
 - superior planet: the planets outside the orbit of the Earth
 - inferior planet: the planets inside the orbit of the Earth

The sidereal period of a superior planet relates to the mean synodic period:

$$ \frac{1}{\rm{sidereal\ period}} + \frac{1}{\rm{synodic\ period}} = 1 $$

The sidereal period of an inferior planet relates to the mean synodic period:

$$ \frac{1}{\rm{sidereal\ period}} - \frac{1}{\rm{synodic\ period}} = 1 $$


Prograde orbits around the sun appear counter-clockwise when viewed from the north.

# 1.4 Kepler's Laws

1. The orbit of each planet is an ellipse; with the Sun at one of its foci
2. Each planet revolves so that the line joining it to the Sun sweeps out equal areas in equal intervals of time.
3. The squares of the periods of any two planets are in the same proportion as the cubes of their mean distances from the Sun. (NB: updates to this law provided in Chapter 6.)

# 1.5 The Astronomical Unit

The mean distance of the Earth from the Sun. Approximately 92,957,000 miles or 149,600,000 km. Convenient that it is of unit value when plugging into the denominator of this relation between the relative periods and distances of planets also orbiting the sun:

$$ \frac{P^2}{P_E^2} = \frac{a^3}{a_E^3} $$

The sidereal year of some planet, designated P, is proportional to its average distance from the Sun, designated a., by this squared-cubed relationship. The corresponding values for the earth are given by PE and AE. (Presumably it should work for any keplerian orbit.)

# 1.8 The Celestial Sphere

The direction of an object is given by two angles that give its position on the celestial sphere.

 - celestial sphere: a spherical shell of arbitrarily large radius on which celestial objects appear projected. Typically centered on the center of the Earth, but can be centered on the observer on Sun, depending on context.
 - ecliptic: the path of the sun on the celestial sphere, inclined to the celestial equator by 23 degrees and 27 minutes
 - equatorial system: a system of spherical coordinates relative to the axis of the Earth's rotation, which defines the north and south poles. The celestial equator is the midpoint between the poles. Positions in the equatorial system are measured in right ascension and declination.
 - right ascension: in the equatorial coordinate system, the angle measured eastward from the prime meridian measured in units of time such that 24 hours equal 360 degrees
 - declination: in the equatorial coordinate system, the angle north (positive) or south (negative) of the celestial equator, measured in degrees
 - celestial longitude: in the ecliptic coordinate system, the longitude is the angle eastward from the prime meridian at the vernal equinox, measured in degrees
 - celestial latitude: in the ecliptic coordinate system, the latitude is the angle north (positive) or south (negative) of the ecliptic, measured in degrees
 - ecliptic system: a system of spherical coordinates relative to the plane of the ecliptic. Positions in the ecliptic system are measured in celestial longitude and latitude, more modernly referred to as ecliptic longitude and latitude. While this is introduced on page 8 of Danby, the actual terms "ecliptic system" or "ecliptic coordinates" do not appear in the text
