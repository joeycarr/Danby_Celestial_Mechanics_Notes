1. Prove that $\nabla r^n = n\vec r r^{n-2}$.

2. Show that the components of $\nabla$ in two-dimensional polar coordinates are

    $$\frac{\partial}{\partial r} \quad \rm{and} \quad \frac{1}{r} \cdot \frac{\partial}{\partial \theta}$$

3. Find the components of $\nabla$ in cylindrical coordinates $(r, \phi, z)$ and in spherical polar coordinates $(r, \theta, \phi)$.

4. If $f$ and $g$ are scalar functions of $\vec r$, show that

    $$ \nabla\left(\frac{f}{g}\right) = \frac{g\nabla f - f\nabla g}{g^2} $$
