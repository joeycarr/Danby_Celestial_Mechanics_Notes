3. Show by similar triangles, or otherwise, that $l(\vec a + \vec b) = l\vec a + l\vec b$.

    <img src="Problems_2.2.3.svg"></img>
    
    We can inspect the diagram above and see that the geometric interpretation of the distributive property of scalar multiplication makes intuitive sense. The angle between $\vec a$ and $\vec a+\vec b $ must be the same as the angle between $2\vec a$ and $2(\vec a + \vec b)$. Likewise, the angle between $\vec a$ and $\vec b$ must be the same as the angle between $2\vec a$ and $2\vec b$. With two angles proven to be the same, the third angle of a triangle must be the same since the interior angles sum to 180 degrees. It follows that the triangles formed by the bisection of the parallelograms in the diagram above must be similar, meaning they have the same angles, but only vary in size. Although the diagram establishes this by example with $l=2$, it's clear than $l$ should be able to have any real value because scalar multiplication of a vector can take any real value, and no real value can impact the angle of the vector (except $l=0$, which produces an undefined angle, but is a degenerate case for a triangle).
    
    ---

    If we decompose $a$ and $b$ into the ordered pairs $(x_a, y_a)$ and $(x_b, y_b)$ respectively, then their sum is the pair $(x_a+x_b, y_a+y_b)$. Scalar multiplication of $\vec a$ by $l$ yields the ordered pair $(l x_a, l y_a)$. Thus
    
    $$ l\vec a + l\vec b = (lx_a, ly_a) + (lx_b, ly_b) = (lx_a+lx_b, ly_a+ly_b) $$
    
    Since we already established that $\vec a + \vec b$ is equivalent to $(x_a+x_b, y_a+y+b)$, it follows by the distributive property that
    
    $$ l(\vec a + \vec b) = l(x_a+x_b, y_a+y_b) = (lx_a+lx_b, ly_a+ly_b) $$
    
    Which shows that $l(\vec a + \vec b)$ is equivalent to the same value as $l\vec a + l\vec b$

7. Find the necessary and sufficent conditions for the sum of three nonzero vectors to vanish.

    In a geometric interpretation of summation, this means that the vectors laid tail to head in sequnce should come back around and point at the origin. Since there are three vectors, we know that they must form a triangle. The vectors will be coplanar, but that's not a necessary or sufficient condition, just a consequence of their triangle-ness. What is necessary is that the sum of any two is equal in magintude and opposite in direction to the third. Or, stated another way, the sum of their resolutions should be zero.

8. Show that

    (a) $\frac{(\vec{OA} + \vec{OB})}{2}$ is the mid-point of $\vec{AB}$;
    
    (b) $\frac{(n\vec{OA} + m\vec{OB})}{(m+n)}$ is the center of mass of masses $n$ at $A$ and $m$ at $B$;
    
    (c) $\frac{(\vec{OA} + \vec{OB} + \vec{OC})}{3}$ is the centroid of the triangle $ABC$.
    
    ---
    
    (a) The midpoint of $\vec{AB}$ is the average of the points $A$ and $B$. This is clear if we think of the midpoint as halfway between the vertical offset between $A$ and $B$ and, independently, halfway between the horizontal offset between $A$ and $B$. We can express this average as:
    
    $$ (\frac{x_a+x_b}{2}, \frac{y_a+y_b}{2}, \frac{z_a+z_b}{2}) $$
    
    If we expand the vector expression $\frac{(\vec{OA} + \vec{OB})}{2}$, we get
    
    $$ \frac{(x_a, y_a, z_a) + (x_b, y_b, z_b)}{2} =  \frac{(x_a+x_b, y_a+y_b, z_a+z_b)}{2} = (\frac{x_a+x_b}{2}, \frac{y_a+y_b}{2}, \frac{z_a+z_b}{2}) $$
    
    (b) With a simple scale having two masses and a balance point, if the ratio of the distances between the masses and the balance point is 2:1, then we can infer that the magnitude of the masses are also in a ratio of 2:1. The center of mass is an average of the positions of the two point masses, weighted by the respective masses. This is the physical interpretation of the equation given in the problem. The quantities $m$ and $n$ are the magnitudes of the masses. If $m = n$, then the center of mass should be equidistant between the masses, and we can see that in such cases, the formula is equivalent to a midpoint as established in (a)
    
    $$ \frac{(n\vec{OA} + m\vec{OB})}{(m+n)} = \frac{(m\vec{OA} + m\vec{OB})}{2m} = \frac{(mx_a, my_a, mz_a)+(mx_b, my_b, mz_b)}{2m} = \frac{m(x_a+x_b,y_a+y_b,z_a+z_b)}{2m} = \frac{(x_a+x_b,y_a+y_b,z_a+z_b)}{2} $$
    
    (c) The centroid of a triangle is defined as the arithmetic mean position of the points in a figure. So the proof here is asking whether one third the sum of these vectors is equivalent to an arithmetic mean. The arithmetic mean $A$ is in turn defined:
    
    $$ A = \frac{1}{n}\sum_{i=1}^{n} a_i $$
    
    Given $a_1=\vec{OA}$, $a_2=\vec{OB}$, and $a_3=\vec{OC}$, and $n=3$, it's clear that
    
    $$ A = \frac{1}{3}(\vec{OA}+\vec{OB}+\vec{OC}) = \frac{\vec{OA}+\vec{OB}+\vec{OC}}{3} $$

11. Express the vector equation of the line (2.2.1; below), in cartesians, and eliminate the parameter $\lambda$.

    $$ \vec r = \vec r_0 + \lambda\uvec i $$
    
    The answer here is a total theft ... [a cheat](https://tutorial.math.lamar.edu/Classes/CalcIII/EqnsOfLines.aspx).

    Note that $\vec r_0$ is a constant, and it is the exapanded values of $\uvec i$, also a constant, that uniquely shapes the slope of the plane. Let's make $\vec r = (x, y, z)$ and $\vec r_0 = (x_0, y_0, z_0)$, and we can assign arbitrary components such that the direction vector $\uvec i = (a, b, c)$.
    
    $$ (x, y, z) = (x_0, y_0, z_0) + \lambda (a, b, c) $$
    
    Since the three vectors' components are all also equal, this gives us three parametric equations:
    
    $$ \begin{align}
    x &= x_0 + \lambda a \\
    y &= y_0 + \lambda b \\
    z &= z_0 + \lambda c
    \end{align} $$
    
    All three equations share the common variable lambda, so we can eliminate lambda by solving for lambda and setting them equal
    
    $$ \frac{x-x_0}{a} = \frac{y-y_0}{b} = \frac{z-z_0}{c} $$

13. If $Ox$ points toward the vernal equinox, and $Oz$ toward the celestial north pole, show that the direction cosines of a point $(\alpha, \delta)$ are

    $$ (\cos \alpha \cos \delta, \quad \sin \alpha \cos \delta, \quad \sin \delta). $$
    
    Find the direction cosines of the following points: the north pole of the ecliptic; the winter solstice; $(12\rm{h}, +45^\circ)$; $(15\rm{h}, -30^\circ)$; $(4\rm{h} \, 35\rm{m} \, 53.2\rm{s}, -58^\circ16'31'')$. In each case verify that the sum of the squares of the direction cosines is one.

```python
import math

def hms(h, m=0, s=0):
    return math.pi * (h/12 + m/720 + s/43200)

def dms(d, m=0, s=0):
    return math.pi * (d/180 + m/10800 + s/648000)

def dircos(alpha, delta):
    '''Find the direction cosines of alpha (right ascension) and delta (declination).'''
    return (
        math.cos(alpha)*math.cos(delta),
        math.sin(alpha)*math.cos(delta),
        math.sin(delta)
    )

def dist(x, y, z):
    return math.sqrt(x**2 + y**2 + z**2)
```

The north pole of the ecliptic

```python
dircos(hms(6), dms(90) + dms(23, 27)), dist(*dircos(hms(6), dms(90) + dms(23, 27)))
```

The winter solstice is dipped 23 degrees 27 minutes below the equator in the "south" relative to the vernal equinox (negative 90 degrees).

```python
dircos(hms(-6), -dms(23, 27)), dist(*dircos(dms(-90), -dms(23, 27)))
```

```python
dircos(hms(12), dms(45)), dist(*dircos(hms(12), dms(45)))
```

```python
dircos(hms(15), dms(-30)), dist(*dircos(hms(15), dms(-30)))
```
