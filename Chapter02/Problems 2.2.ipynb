2. State the necessary and sufficient conditions for the sum of two vectors to vanish.

    The two vectors must be of the same length (magnitude) and opposite direction. Since negation is geometrically equivalent to a 180 degree turn in direciton, opposite direction can arise either by switching endpoints or negating the vector.

3. Show by similar triangles, or otherwise, that $l(\vec a + \vec b) = l\vec a + l\vec b$.

    If we decompose $a$ and $b$ into the ordered pairs $(x_a, y_a)$ and $(x_b, y_b)$ respectively, then their sum is the pair $(x_a+x_b, y_a+y_b)$. Scalar multiplication of $\vec a$ by $l$ yields the ordered pair $(l x_a, l y_a)$. Thus
    
    $$ l\vec a + l\vec b = (lx_a, ly_a) + (lx_b, ly_b) = (lx_a+lx_b, ly_a+ly_b) $$
    
    Since we already established that $\vec a + \vec b$ is equivalent to $(x_a+x_b, y_a+y+b)$, it follows by the distributive property that
    
    $$ l(\vec a + \vec b) = l(x_a+x_b, y_a+y_b) = (lx_a+lx_b, ly_a+ly_b) $$
    
    Which shows that $l(\vec a + \vec b)$ is equivalent to the same value as $l\vec a + l\vec b$

4. Show that the resolution of a vector along any three directions (where no two are parallel or antiparallel and all three do not lie in a plane) is unique. (Consider the consequences if this were not so).

    The vector composed of the ordered triplet $(x, y, z)$ is trivial to resolve to $x\uvec i + y\uvec j + z\uvec k$. If this vector somehow had multiple resolutions, that would imply that it refered to a point that had two positions along an axis, i.e. that the axes are looped. If we can take it as _given_ that any three basis vectors that are not parallel or coplanar are a linear transform of an orthonormal basis, then the same reasoning applies to the resolution along the transformed directions, just the same as the cartesian axes. (I don't feel like this is a satisfactory answer because that's a big _given_.)
    
    Maybe another way to think about it is in terms of infinitesimal wiggles. If we wiggle the coefficient of one of the basis vectors, then the vector moves even though the other two coefficients remain the same. There shouldn't be any way to wiggle the other coefficients to compensate for the displacement along the first 'axis', because they're independent. How to phrase that as a proof?

5. Find the moduli and direction cosines of the following vectors:

    $$ (-a, 0, 0), \quad (1, 1, -1), \quad (0, 1, -1), \quad (x, y, z) $$
    
    and of
    
    $$ [(-a, b, 0) + (a, b, 0)], \quad [(2, 3, -5) + (3, -2, 4)] $$
    
    and 
    
    $$ [(x, y, z) + (p, q, r)]. $$
    
    All symbols may be assumed to represent positive numbers.
    
    See p. 17-18 for the definition of the direction cosine. There's one for each axis, and it's the projection on that axis divided by the magnitude of the vector.
    
    $$ (x, y, z) = (rl, rm, rn) $$
    
    If we devide through by $r$ we get
    
    $$ (l, m, n) = (\frac{x}{r}, \frac{m}{r}, \frac{n}{r}) $$
    
    So moduli and direction cosines of the following:
    
    $$ \Vert(-a, 0, 0)\Vert = \sqrt{(-a)^2 + 0^2 +0^2} = \vert a\vert;\quad l = \frac{-a}{\Vert a\Vert} = -1,\ m = 0,\ n = 0 $$
    
    $$ \Vert(1, 1, -1)\Vert = \sqrt{1^2+1^2+(-1)^2} = \sqrt{3};\quad (l, m, n) = \left(\frac{1}{\sqrt{3}}, \frac{1}{\sqrt{3}}, \frac{-1}{\sqrt{3}}\right) $$
    
    $$ \Vert(x, y, z)\Vert = \sqrt{x^2+y^2+z^2};\quad (l, m, n) = \left(\frac{x}{\sqrt{x^2+y^2+z^2}}, \frac{y}{\sqrt{x^2+y^2+z^2}}, \frac{z}{\sqrt{x^2+y^2+z^2}}\right) $$
    
    and of
    
    $$ [(-a, b, 0) + (a, b, 0)] = (-a+a, b+b, 0+0) = (0, 2b, 0) $$
    
    $$ \Vert(0, 2b, 0)\Vert = 2\vert b\vert; \quad (l, m, n) = (0, \frac{2b}{2\vert b\vert}, 0) = (0, 1, 0) $$
    
    $$ [(2, 3, -5) + (3, -2, 4)] = (5, 1, -1) $$
    
    $$ \Vert(5, 1, -1)\Vert = \sqrt{5^2, 1^2, (-1)^2} = \sqrt{25+1+1} = \sqrt{27} $$
    
    $$ (l, m, n) = \left(\frac{5}{\sqrt{27}}, \frac{1}{\sqrt{27}}, \frac{-1}{\sqrt{27}}\right) \approx (0.9623, 0.1925, -0.1925) $$
    
    and
    
    $$ [(x, y, z) + (p, q, r)] = (x+p, y+q, z+r) $$
