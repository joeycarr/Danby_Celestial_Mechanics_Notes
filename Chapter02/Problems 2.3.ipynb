# Problems 2.3

$$ \newcommand{\uvec}[1]{\hat{\mathbf{#1}}} $$

1. If the angle between $\vec{AB}$ and $\vec{CD}$ is $\theta$, what are the angles between $\vec{AB}$ and $\vec{DC}$ and between $\vec{BA}$ and $\vec{DC}$?

2. Find a formula for the cosine of the angle between the two vectors $(x_1, y_1, z_1)$ and $(x_2, y_2, z_2)$.

3. Prove by a vector method thatfor any triangle

    $$ a^2 = b^2 + c^2 - 2bc\cos A $$

4. Solve, for the unknown vector $\vec x$, the equations:

    (a) $\vec x \cdot \vec a = 0$
    
    (b) $\vec x \cdot \vec a = \vec x \cdot \vec b$ (where $\vec a \ne \vec b$)
    
    (c) $\vec x \cdot \vec a = b$

5. Evaluate $(\vec a + \vec b) \cdot (\vec a - \vec b)$ when $a = b$. Hence prove that the diagonals of a rhombus are perpendicular.

6. $\vec a$, $\vec b$, and $\vec c$ are three nonzero and co-planar vectors, and no two of them are parallel or antiparallel. Show that if

    $$ \vec x \cdot \vec a = \vec x \cdot \vec b = \vec x \cdot \vec c = 0 $$

    then $\vec x = 0$.

7. Assuming that forces are vectors, what relation must exist between $P$ and $Q$ such that the forces $P$ along $(1, 0, 0)$ and $Q$ along $(0, 1, 0)$ have zero resultant along $(l, m, 0)$?

8. Show by a vector method that the altitudes of a triangle intersect in a point.

9. Show by a vector method that the perpendicular bisectors of a triangle intersect in a point.

10. Show that the equation of a plane can be written

    $$ (\vec r - \vec r_0) \cdot \uvec i = 0 $$
    
    Interpret $\uvec i$, and show that this equation is equivalent to that of problem 12, Section 2.2.

11. Find the shortest distances from the origin to the line

    $$ \vec r = \vec r_0 + \lambda\uvec i $$
    
    and to the plane
    
    $$ (\vec r - \vec r_0) \cdot \uvec i = 0 $$

12. Find the shortest distances from the point $P_1$, where $\vec{OP_1} = \vec r_1$, to the line and plane of problem 11.

13. Interpret the equation

    $$ (\vec r - \vec r_0)^2 = a^2 $$
    
    where $\vec r$ is a variable vector.
