2. Find a formula for the cosine of the angle between the two vectors $(x_1, y_1, z_1)$ and $(x_2, y_2, z_2)$.

    Given that
    
    $$ \vec A \cdot \vec B = \Vert\vec A\Vert\ \Vert\vec B\Vert\cos{\theta} $$
    
    And that
    
    $$ \vec A \cdot \vec B = x_A x_B + y_A y_B + z_A z_B $$
    
    If we set $\vec A = (x_1, y_1, z_1)$ and $\vec B = (x_2, y_2, z_2)$, then we have
    
    $$ \Vert\vec A\Vert\ \Vert\vec B\Vert\cos{\theta} = x_1 x_2 + y_1 y_2 + z_1 z_ 2 $$
    
    Dividing both sides by the product of the magnitudes of $\vec A$ and $\vec B$, we have
    
    $$ \cos \theta = \frac{x_1 x_2 + y_1 y_2 + z_1 z_ 2}{\Vert\vec A\Vert\ \Vert\vec B\Vert} $$
    
    Which we can fully expand to
    
    $$ \cos \theta = \frac{x_1 x_2 + y_1 y_2 + z_1 z_ 2}{\sqrt{x_1^2 + y_1^2 + z_1^2}\sqrt{x_2^2 + y_2^2 + z_2^2}} $$
    
    Or, if we want to retain vector form, we might never have expanded the numerator
    
    $$ \cos \theta = \frac{\vec A \cdot \vec B}{\Vert\vec A\Vert\ \Vert\vec B\Vert} $$

3. Prove by a vector method that for any triangle

    $$ a^2 = b^2 + c^2 - 2bc\cos A $$
    
    This makes sense when $A = \pi / 2$, then $\cos A = 0$ and the third term is eliminated; the result is something that looks very much like the pythagorean theorem, although the hypotenuse is labeled $a$ rather than $c$.
    
    <img src="Problems_2.3.3.svg" style="max-width: 70%;" />
    
    We'll invent vectors $\vec a, \vec b, and \vec c$ that have length $a = \Vert\vec a\Vert$, $b = \Vert\vec b\Vert$, and $c = \Vert\vec c\Vert$. Each angle, $A$, $B$, and $C$ is opposite the side with the corresponding lowercase letter. I imagine that Danby expects us to be familiar with this formula, which goes by the name "the law of cosines", and to recall that this is the canonical labeling.
    
    I "cheated" and just looked up the answer on Wikipedia, which has a large listing of proofs of the law of cosines, including one [using vectors](https://en.wikipedia.org/wiki/Law_of_cosines#Using_vectors).
    
    We start by noting that 
    
    $$ \vec a = \vec c - \vec b $$
    
    Since we can take the dot product of each side with itself and continue to have a balanced equation, analogous to squaring each side of an equation, we can rewrite this identity for $\vec c$ as
    
    $$ \vec a \cdot \vec a = (\vec b - \vec c)\cdot(\vec b - \vec c) $$
    
    Recall the defining identity of the dot product, which we'll restate here using $\vec u$ and $\vec v$ to avoid confusion
    
    $$ \vec u \cdot \vec v = \Vert\vec u\Vert\ \Vert\vec v\Vert \cos\theta $$
    
    When the dot product is taken with itself, the $\cos \theta$ term is $\cos(0)$, which equals $1$, so the result is the magnitude of the input vector squared. We can expand the right-hand side like any polynomial multiplication, using the _first inner outer last_ (FOIL) mnemonic. This allows us to restate the above as
    
    $$ \Vert\vec a\Vert^2 = \Vert\vec b\Vert^2 + \Vert\vec c\Vert^2 + 2(\vec b \cdot \vec c) $$
    
    Or more simply using the scalars that correspond to the similarly labeled vectors:
    
    $$ a^2 = b^2 + c^2 + 2(\vec b \cdot \vec c) $$
    
    We still have an unresolved dot product, but again we can apply the defining identity of the dot product to obtain
    
    $$ a^2 = b^2 + c^2 + 2(\Vert\vec b\Vert\ \Vert\vec c\Vert\cos A) $$
    
    Again, subsitituting the simple scalars for the magnitude expressions, we finally obtain the expression that Danby asked us to prove
    
    $$ a^2 = b^2 + c^2 + 2bc\cos A $$

(a) $\vec x \cdot \vec a = 0 $

If we take it as given that a plane can be defined implicitly as 

$$ Ax + By + Cz + D = 0 $$

Then we can see that the expansion of the dot product looks like a plane. Here the coefficients are $A = a_x$, $B = a_y$, $C = a_z$, and $D = 0$.

$$ \vec x \cdot \vec a = x_x a_x + x_y a_y + x_z a_z = 0 $$

    \end{align}

If we go with geometric algebra instead, we can instead add $\vec x \wedge \vec a$ to both sides

(c) $\vec x \cdot \vec a = b$

Let's get its expansion in terms of lengths and angles.

$$ b = \Vert\vec x\Vert \Vert\vec a\Vert \cos \theta $$

If we solve for the length of $\vec x$ as a function of the cos, we can make a straightforward graph in polar coordinates using $\theta$ as the independent variable.

$$ \Vert \vec x \Vert = \frac{b}{\Vert \vec a \Vert \cos \theta} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.arange(-np.pi, np.pi, 0.01)
b = 1
a = 2
r = b / (a * np.cos(theta))

x = r*np.cos(theta)
y = r*np.sin(theta)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(True)
plt.show()
```

So the interpretation here is that $\vec x \cdot \vec a = b$ has infinitely many solutions and that $\vec x$ varies on a line perpendicular to $\vec a$. Thus in a 3D context, the equation describes a cylinder with radius $b / \Vert\vec a\Vert$.

$$ (a_x+b_x)(a_x-b_x) + (a_y+b_y)(a_y-b_y) + (a_z+b_z)(a_z-b_z) $$

We can regonize a factorization of a difference of squares, which makes the next step slightly easier

$$\begin{align}

a_x^2 + a_y^2 + a_z^2 &= b_x^2 + b_y^2 + c_y^2
\end{align}$$

Which basically says that the lengths $a = b$.
