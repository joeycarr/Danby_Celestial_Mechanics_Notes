1. Find expressions for the modulus and direction of the angular velocity of the mean Sun.

2. Find the angular velocity of the apparent Sun, assuming that the Earth's orbit is circular, with constant angular velocity. Resolve this angular velocity along an equitorial system of axes and hence find the components of the Sun's motion on the celestial sphere along these axes. Compare the component in right ascension with the angular velocity of the mean Sun.

3. Assuming the Earth to be fixed and the celestial sphere (of unit radius) to be rotating around the Earth, consider the apparent motion of the stars seen by an observer at latitude $l$. If $\uvec z$ points to the observer's zenith (the direction vertically above her), resolve the motion of a star at $\delta$ along and at right angles to $\uvec z$.
