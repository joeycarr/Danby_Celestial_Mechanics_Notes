# Problems 2.4

$$ \newcommand{\uvec}[1]{\hat{\mathbf{#1}}} $$

1. Let $\uvec i$ point torward the vernal equinox, $\uvec k$ toward the celestial north pole, $\uvec m$ toward the summer solstice, and $\uvec n$ toward the north pole of the ecliptic. Express $\uvec i$ in terms of $\uvec n$ and $\uvec k$; $\uvec m$ in terms of $\uvec n$ and $\uvec k$; $\uvec n$ in terms of $\uvec i$ and $\uvec k$; $\uvec k$ in terms of $\uvec i$ and $\uvec n$.

2. Prove the formula for the triple vector product.

3. If $\vec a + \vec b + \vec c = 0$, show that 

    $$ \vec b \times \vec c = \vec c \times \vec a = \vec a \times \vec b $$

    and interpret geometrically.

4. What is the locus of $\vec r$ such that $[\vec r, \vec a, \vec b] = 0$?

5. If $l$, $m$, and $n$ are three nonzero scalars, and 

    $$ l\vec a + m\vec b + n\vec c = 0 $$
    
    show that $\vec a$, $\vec b$, and $\vec c$ are coplanar. If also
    
    $$ l'\vec a + m'\vec b + n'\vec c = 0 $$
    
    show that
    
    $$ \frac{l}{l'} = \frac{m}{m'} = \frac{n}{n'} $$

6. If $\vec x$, $\vec a$, and $\vec b$ are coplanar, show that usually $\vec x$ can be expressed as

    $$ \vec x = \lambda\vec a + \mu\vec b $$
    
    What is the exception to this?

7. Evaluate $(\vec a + \vec b) \times (\vec a - \vec b)$.

8. Choosing simple (but not trivial) numerical values for the components of $\vec a$, $\vec b$, and $\vec c$, verify numerically that

    $$ \vec a \cdot (\vec b + \vec c) = \vec a \cdot \vec b + \vec a \cdot \vec c, \quad \vec a \times (\vec b + \vec c) = \vec a \times \vec b + \vec a \times \vec c, \\
    \vec a \times (\vec b \times \vec c) = (\vec a \cdot \vec c)\vec b - (\vec a \cdot \vec b)c $$
    
    Find the values of
    
    $$ |\vec a \times \vec c|, \quad [\vec a, \vec b, \vec c], \quad (\vec a \times \vec b) \times \vec c, \quad (\vec a \times \vec b) \cdot (\vec a \times \vec c), \quad (\vec a \times \vec b) \times (\vec a \times \vec c) $$

9. Find the equations to the plane containing the heads of the vectors $\vec a$, $\vec b$, and $\vec c$.

10. Show that $(\vec a \times \vec b) \cdot (\vec c \times \vec d) = (\vec b \cdot \vec d)(\vec c \cdot \vec a) - (\vec b \cdot \vec c)(\vec d \cdot \vec a)$.

11. Show that

    $$ \begin{align}(\vec a \times \vec b) \times (\vec c \times \vec d)
        &= [\vec a,\vec b, \vec d]\vec c - [\vec a, \vec b, \vec c]\vec d \\
        &= (\vec c \times \vec d) \times (\vec b \times \vec a) \\
        &= [\vec c, \vec d, \vec a]\vec d - [\vec c, \vec d, \vec b]\vec a
        \end{align}
    $$
    
    Hence show that
    
    $$ [\vec a, \vec b, \vec c]\vec d = [\vec d, \vec b, \vec c]\vec a + [\vec a, \vec d, \vec c]\vec b + [\vec a, \vec b, \vec d]\vec c $$
    
    and deduce the conditions for the unique resolution of the vector $\vec d$ along the three directions $\vec a$, $\vec b$, and $\vec c$.

12. Show that

    $$ [\vec a, \vec b, \vec c](\vec f \times \vec g) =
        \left| \begin{matrix}
            \vec a & \vec b & \vec c \\
            \vec f \cdot \vec a & \vec f \cdot \vec b & \vec f \cdot \vec c \\
            \vec g \cdot \vec a & \vec g \cdot \vec b & \vec g \cdot \vec c
        \end {matrix} \right|
    $$
    
    Hence, generalize equation (2.4.2)
    
    $$ \vec b \times \vec c = \left| \begin{matrix}
            \uvec i & \uvec j & \uvec k \\
            b_x & b_y & b_z \\
            c_x & c_y & c_z
        \end{matrix} \right|
    $$

13. Solve for $\vec X$ the equation $\vec X \times \vec a = \vec b$, where $\vec a \cdot \vec b = 0$.

14. Solve the equation $\vec X + \vec a(\vec X \cdot \vec b) = \vec c$. (Try $\cdot \vec b$.)

15. Solve the simultaneous equations $\alpha \vec X + \beta \vec Y = \vec a$ and $\vec X \times \vec Y = \vec b$ where $\vec a \cdot \vec b = 0$.

16. Solve the equation $\vec X \times \vec a + (\vec X \cdot \vec b)\vec c = \vec d$
