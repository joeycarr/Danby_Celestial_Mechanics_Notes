2. Prove the formula for the triple vector product.

The formula of the triple vector product is:

$$ \vec a \times (\vec b \times \vec c) = (\vec a \cdot \vec c)\vec b - (\vec a \cdot \vec b)\vec c $$

Let's simply try verifying it by expanding both expressions algebraically.

```python
import warnings
from sympy import *
init_printing()
from sympy.vector import CoordSys3D, Vector

warnings.filterwarnings("ignore", category=UserWarning)
N = CoordSys3D('N')
```

```python
a = symbols('a:3')
b = symbols('b:3')
c = symbols('c:3')
```

```python
av = a[0]*N.i + a[1]*N.j + a[2]*N.k
bv = b[0]*N.i + b[1]*N.j + b[2]*N.k
cv = c[0]*N.i + c[1]*N.j + c[2]*N.k
```

```python
tripleCross = av.cross(bv.cross(cv))
tripleCross
```

```python
tripleCrossEquiv = (av.dot(cv))*bv - (av.dot(bv))*cv
tripleCrossEquiv
```

```python
diff = tripleCross - tripleCrossEquiv
diff
```

```python
simplify(diff)
```

7. Evaluate $(\vec a + \vec b) \times (\vec a - \vec b)$.

We can expand it like a polynomial, eliminate doubles, and subsitute the negatives for the reversed.

$$\begin{align}
(\vec a + \vec b) \times (\vec a - \vec b) &= \vec a \times \vec a - (\vec a \times \vec b) + (\vec b \times \vec a) - (\vec b \times \vec b) \\
&= 0 - (\vec a \times \vec b) + (\vec b \times \vec a) - 0 \\
&= (\vec b \times \vec a) + (\vec b \times \vec a) \\
&= 2(\vec b \times \vec a)
\end{align}$$

Since we've demonstrated the equivalencies via brute force algebra (hidden in Sympy code), we can take the following given equivalency as axiomatic:

$$ [\vec a,\vec b, \vec d]\vec c - [\vec a, \vec b, \vec c]\vec d = 
   [\vec c, \vec d, \vec a]\vec b - [\vec c, \vec d, \vec b]\vec a
$$

So then we have

$$ \begin{align}
    - [\vec a, \vec b, \vec c]\vec d &= [\vec c, \vec d, \vec a]\vec b - [\vec c, \vec d, \vec b]\vec a - [\vec a,\vec b, \vec d]\vec c \\
    [\vec a, \vec b, \vec c]\vec d &= - [\vec c, \vec d, \vec a]\vec b + [\vec c, \vec d, \vec b]\vec a + [\vec a,\vec b, \vec d]\vec c \\
    &= [\vec c, \vec d, \vec b]\vec a - [\vec c, \vec d, \vec a]\vec b + [\vec a,\vec b, \vec d]\vec c \\
\end{align} $$

Then we can roll the order of the vectors to the left inside the brackets without impacting the magnitude of the triple scalar product. Thus

$$ \begin{align}
    [\vec a, \vec b, \vec c]\vec d &= [\vec c, \vec d, \vec b]\vec a - [\vec c, \vec d, \vec a]\vec b + [\vec a,\vec b, \vec d]\vec c \\
    &= [\vec d, \vec b, \vec c]\vec a + [\vec a, \vec d, \vec c]\vec b + [\vec a,\vec b, \vec d]\vec c
\end{align} $$
